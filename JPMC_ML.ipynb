```python

from keras.models import Sequential
from keras.layers import LSTM, Dense, Bidirectional, LeakyReLU,Dropout
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
from keras.losses import mean_squared_error
from keras.metrics import RootMeanSquaredError

def data_reshape(x_train, y_train, n_steps, n_features, X, y):
    # X = []
    # for i in np.arange(n_steps, len(data)):
    #     X.append(data.iloc[i-n_steps:i, :])
    # X = np.array(X)
    # X = np.reshape(X, (X.shape[0], n_steps, n_features))
    for i in np.arange(n_steps, len(x_train)):
        X.append(x_train.iloc[i-n_steps:i, :])
        y.append(y_train.iloc[i, :])
    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], n_steps, n_features))
    return X, y

# Define the trading strategy function


def trading_strategy(model, X,y_pred):
    y_pred = model.predict(X)
    return y_pred


def lstm_model(model):
    model.add(LSTM(units=120, return_sequences=True, input_shape=(n_steps, n_features), use_bias=True,activation='relu'))
    model.add(Dropout(0.1))
    model.add(LSTM(units=420, use_bias=True,return_sequences=False,activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(units=200, activation='relu'))
    model.add(Dense(units=1, activation='relu'))
    model.compile(optimizer='adam', loss=mean_squared_error, metrics=[RootMeanSquaredError()])

    return model



def calculate_market_indicators(data):
    # Calculate debt-to-equity ratio
    data['debt_to_equity_ratio'] = data['liabilities'] / data['equity']
    # Calculate return on equity
    data['return_on_equity'] = data['net_income'] / data['equity']

    # Calculate return on assets
    data['return_on_assets'] = data['net_income'] / data['total_assets']

    # Calculate current ratio
    data['current_ratio'] = data['current_assets'] / data['current_liabilities']

    # Calculate earnings per share
    data['eps'] = data['net_income'] / data['shares_outstanding']

    # Calculate price-to-earnings ratio
    # data['p_e_ratio'] = data['price'] / data['eps']

    # Calculate dividend yield
    # data['dividend_yield'] = data['dividend'] / data['price']
    data_stock=data['Stock']
    data.drop('Stock',inplace=True,axis=1)
    for col in data.columns:
      X_scaled = scaler.fit_transform(data[col].values.reshape(-1, 1))
      data[col] = X_scaled.flatten()  # convert the 2D array back to 1D array
    data['Stock']=data_stock
    return data
    # print(indicators_df.head())

from sklearn.preprocessing import LabelEncoder
def labelencoder(df):
# Create sample data
  data = df['Stock']

  # Create a LabelEncoder object
  le = LabelEncoder()

  # Fit and transform the data
  data_encoded = le.fit_transform(data)
  df['Stock']=data_encoded
  return df



# Print the encoded data
# print(data_encoded)
import pandas as pd
import numpy as np
df=pd.read_csv('/Train.csv')
df.set_index('Date', inplace=True)
df=labelencoder(df)
# Define the number of time steps and features for the LSTM model
df=calculate_market_indicators(df)
n_steps = 5
n_features = df.columns.shape[0]
# df.head()



model = Sequential()
model = lstm_model(model)
# Train the LSTM model
X_train = df[df.columns].iloc[:1400,:]
# y_train = data_model[['Open', 'Close']]
y_train = df[['price']].iloc[:1400,:]
# X_train.dropna(inplace=True,axis=0)
# y_train.dropna(inplace=True,axis=0)
X = []
y = []
X, y = data_reshape(X_train, y_train, n_steps, n_features, X, y)

X_val = df[df.columns].iloc[1400:,:]

# y_train = data_model[['Open', 'Close']]
y_val = df[['price']].iloc[1400:,:]
X_v=[]
y_v=[]
X_v, y_v = data_reshape(X_val, y_val, n_steps, n_features, X_v, y_v)
history = model.fit(X, y, epochs=20, batch_size=10,validation_data=(X_v, y_v))

df_test=pd.read_csv('/Test_x.csv')
df_test=calculate_market_indicators(df_test)
df_test.set_index('Date', inplace=True)
# Use the LSTM model to generate trading signals
df_test=labelencoder(df_test)
# Define the number of time steps and features for the LSTM model
df=calculate_market_indicators(df)
X_test = df_test[df_test.columns]
X_t=[]
X_t=data_reshape_test(X_test,n_steps,n_features,X_t)
def data_reshape_test(x_test, n_steps, n_features, X):
    # X = []
    # for i in np.arange(n_steps, len(data)):
    #     X.append(data.iloc[i-n_steps:i, :])
    # X = np.array(X)
    # X = np.reshape(X, (X.shape[0], n_steps, n_features))
    for i in np.arange(n_steps, len(x_test)):
        X.append(x_test.iloc[i-n_steps:i, :])
        # y.append(y_train.iloc[i, :])
    X = np.array(X)
    X = np.reshape(X, (X.shape[0], n_steps, n_features))
    return X

y_pred = []
y_pred = trading_strategy(model, X_t,y_pred)

```
